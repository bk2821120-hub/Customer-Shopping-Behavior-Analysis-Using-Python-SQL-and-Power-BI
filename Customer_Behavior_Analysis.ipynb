```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median))
```

```python
df.isnull().mean()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

create a columns age_group

```python
labels=["Young Adult","Adult","Middle.aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df["age_group"]
```

```python
df['frequency_of_purchases'].describe()
```

```python
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90}
df["frequency_purchases_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Create Engine (this is enough — no psycopg2 needed)
engine = create_engine("postgresql://postgres:admin@localhost:5432/customer")

df_clean = df.copy()

# Remove callable values if any
for col in df_clean.columns:
    df_clean[col] = df_clean[col].apply(lambda x: x() if callable(x) else x)

# Automatically create + insert into PostgreSQL table
df_clean.to_sql(
    "customer_shopping_behavior",
    engine,
    if_exists="replace",   # replace / append / fail
    index=False
)

print("Table created successfully!")
```
